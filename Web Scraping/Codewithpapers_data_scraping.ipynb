```
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

req = Request('https://paperswithcode.com/search?q=nlp', headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage)
```

```
print (soup.prettify())
```

### Titles

```
# for extracting the titles from the paperswithcode
titles = []
x = soup.findAll('h1')
for node in x:
    titles.append(''.join(node.findAll(text=True)))
titles
```

### Papers_links

```
papers = []
x = soup.findAll('div',attrs={'class':'entity'})
for i in x:
  papers.append( i.find('a',attrs={'class':'badge badge-light'})['href'])
papers
```

```
base_url = 'https://paperswithcode.com'
papers_links = []
for i in range(len(papers)):
  papers_links.append('https://paperswithcode.com{}'.format(papers[i]))
papers_links
```

```
def cop_desc():
  desc = []
  x = soup.findAll('div',attrs={'class':'col-lg-9 item-content'})
  for i in x:
    desc.append( i.find('p',attrs={'class':'item-strip-abstract'}).text)
  return desc
```

```
def cop_titles():
  titles = []
  x = soup.findAll('h1')
  for node in x:
      titles.append(''.join(node.findAll(text=True)))
  titles1 = titles[1:] # Because of the first content is coming as a str "Search Result"
  return titles1
```
